```python
#LOCATE JSON FILES IN DIRECTORY
path = '/Volumes/Sarp Akar/data'
json_pattern = os.path.join(path,'*.json')
file_list = glob.glob(json_pattern)
```

```python
#READ JSON FILES -> CONVERT TO PANDAS DATA FRAME
def read_json(file_path):
    df = pd.read_json(file_path, lines= True,orient ='records')
    return df
```

```python
# FILTER & ADD TWEET ATTRIBUTES TO MYSQL
tweet_attributes = pd.DataFrame()
error_files = []
for i in range(len(file_list)):
    print(file_list[i])
    try:
        df = read_json(file_list[i])
        ext_text = df['extended_tweet'].apply(pd.Series)['full_text'].to_frame()
        df['text']= ext_text['full_text'].fillna(df['text'])
        df = df[['created_at', 'text', 'id_str', 'in_reply_to_status_id_str', 'in_reply_to_user_id_str','lang']]
        tweet_attributes = tweet_attributes.append(df)
    except:
        error_files.append(i)
```

```python
#ADD USER ATTRIBUTES TO MYSQL
for i in range(len(file_list)):
    try:
        df = read_json(file_list[i])
        df_user_data = df['user']
        df_normalize= pd.json_normalize(json.loads(df_user_data.to_json(orient = 'records')))
        df_user_att = df_normalize['id_str'].to_frame()
        df_user = df[['created_at','id_str']].copy()
        df_user['user_id'] = df_user_att
        df_user.to_sql(name = 'users_test', con = engine, index = False, if_exists = 'append', chunksize = 1000, method = 'multi')
        print(file_list[i])
    except:
        print('error')
```

```python
#ADD PLACE ATTRIBUTES TO MYSQL
count = 0
for i in range(len(file_list)):
    try:
        place = read_json(file_list[i])['place']
        place = place.apply(lambda x: {} if pd.isna(x) else x)
        place = pd.json_normalize(json.loads(place.to_json(orient = 'records'))).dropna()
        place = place[['country_code']]
        place['tweet_id'] = read_json(file_list[i])['id_str']
        place.to_sql('place', con = engine, index = False, if_exists = 'append')
        print(file_list[i])
    except:
        print('error')
```
